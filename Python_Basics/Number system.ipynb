# how to swap to varaible in python

```python
# how to swap to varaible in python
a=5
b=6
# a,b=b,a
```

```python
print(a,b)
```

```python
temp=a
a=b
b=temp
print(a,b)
```

```python
a2=5
b2=6
a2=a2+b2
b2=a2-b2
a2=a2-b2
```

```python
print(a2,b2)
```

```python
a,b=b,a
print(a,b)
```
