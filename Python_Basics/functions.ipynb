# variable length argument
- it can accept mutiple argument

```python
# variable length argument
def sum(a,b):
    c=a+b
    return c
sum(5,6)
```

```python
def sum(a,b):
    c=a+b
    return c
sum(5,6,7,8)
```

```python
def sum(a,*b):
    c=a+b
    return c
sum(4,5,2,5) # int not concat with tuple
```

```python
def sum(a,*b):
    # c=a+b
    print(a)
    print(type(a))

    print(b)
    print(type(b))
sum(5,3,56,4)
```

```python
## 1st argument is fixed & we fetch each value from the tuple & we can add them. 

def sum(a,*b):
    c=a
    for i in b:
        c=c+i
    print(c)
sum(5,2,6,7)
```

```python
def sum(*a,*b): # we can not assign two keyword argument
    c=a
    for i in b:
        c=c+i
    print(c)
sum(4,5,2,4,3,2,4,-2,24)
```

```python
def sum(*a,c):
    print(a)
    print(c)
    for i in a:
        print(i+c)
sum(2,3,4,55,2,4) # in variable length argument we have to assign keyword argument if the variable length arg assign at first
```

```python
def sum(*a,c):
    print(a)
    print(c)
    for i in a:
        print(i+c)
sum(2,3,4,55,2,4,c=50)
```

# **kwargs
- its takes n numbers of keyword argument without defined in formal argument
- its defined at the time of function calling
- we need to takes a variable for each argument
- its return the dictionary values

```python
def sum(**kwargs):
    print(kwargs)
sum(a=(23,22),b=3,c=4)
```

```python
def sum(**kwargs):
    print(kwargs)
sum(1,3,3)
```

```python
def sum(**kwargs):
    print(kwargs)
sum(a=1,3,b=3) #we have to give keyword for each agrumnet
```
