```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

# read and display an image with OpenCV+matplotlib

```python
# Read an image with OpenCV (BGR format)
# OpenCV bydefault read image in BGR color format whereas numpy read in RGB color format
img_path=r'C:\Users\USER\Desktop\dog1.jpg'
image=cv2.imread(img_path) 
```

```python
print(type(image))
print(image.shape) # it shows height ,width and dimensions 
```

```python
# will look color-shifted (BGR  vs RGB)
plt.imshow(image)
plt.show()
```

```python
plt.imshow(image)
plt.axis('off') # axis is not shown
plt.show()
```

```python
# convert BGR to RGB
fixed_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
print(fixed_image.shape)
print(image.shape)
```

```python
plt.imshow(fixed_image)
plt.axis('off')
plt.show()
```

# Read GrayScale And Change colormaps

```python
img_gray=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
print(img_gray.shape)
print(img_gray.min(),img_gray.max())
```

```python
# Default grayscale display
plt.imshow(img_gray)
plt.axis('off')
plt.show()
```

```python
# Explicit gray cmap
plt.imshow(img_gray,cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Another colormap, e.g. 'magma' or 'accent' or many more check the cmaps
# Valid colormaps like gray, magma, viridis, plasma, Blues, Accent are mentioned in the session
plt.imshow(img_gray,cmap='magma')
plt.axis('off')
plt.show()
```

# Resize image (fixed size and ratio)

```python
# original shape
print(image.shape)
```

```python
# Resize to explicit size (height=500, width=367)
resized=cv2.resize(image,(367,500))
print(resized.shape)
```

```python
plt.imshow(cv2.cvtColor(resized,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Resize using width/height ratio(50%)
w_ratio=0.5
h_ratio=0.5
h,w=image.shape[:2]
new_w=int(w*w_ratio)
new_h=int(h*h_ratio)

resized1=cv2.resize(image,(new_w,new_h))
print(resized1.shape)

plt.imshow(cv2.cvtColor(resized1,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

#  Flip image and save new image

```python
# Flip image (vertical flip:flipCode=0)
img_flip0=cv2.flip(image,0)
# Horizonatal flip:flipcode=1
img_flip1=cv2.flip(image,1)
# Both axes: flipCode=-1
img_flip2=cv2.flip(image,-1)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1);plt.imshow((img_flip0));plt.title('flip 0 vertical');
plt.subplot(1,3,2);plt.imshow((img_flip1));plt.title('flip 1 horizontal');
plt.subplot(1,3,3);plt.imshow((img_flip2));plt.title('flip -1 sides changes');
plt.show()
```

```python
# Flip image (vertical flip: flipCode=0)
img_flip0 = cv2.flip(image, 0)

# Horizontal flip: flipCode=1
img_flip1 = cv2.flip(image, 1)

# Both axes: flipCode=-1
img_flip2 = cv2.flip(image, -1)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(cv2.cvtColor(img_flip0, cv2.COLOR_BGR2RGB)); plt.title("flip 0"); plt.axis("off")
plt.subplot(1,3,2); plt.imshow(cv2.cvtColor(img_flip1, cv2.COLOR_BGR2RGB)); plt.title("flip 1"); plt.axis("off")
plt.subplot(1,3,3); plt.imshow(cv2.cvtColor(img_flip2, cv2.COLOR_BGR2RGB)); plt.title("flip -1"); plt.axis("off")
plt.show()

# Save one flipped image
# cv2.imwrite("new_gen_ai_image.jpg", img_flip2)
```

# Webcame video capture: grayscale and RGB
# read the video using opencv

```python
cap=cv2.VideoCapture(0) # 0=default  webcam

# Optionally get frame width/height
frame_width =cap.get(cv2.CAP_PROP_FRAME_WIDTH)
frame_height=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
print(frame_width,frame_height)

while True:
    ret,frame =cap.read()

    if not ret:
        break

    # Example 1: grayscale video
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('Gray Video',gray_frame)

    # Quit on 'q'
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
    
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    cv2.imshow("RGB Video", rgb_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2Luv)
    cv2.imshow("RGB Video", rgb_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
