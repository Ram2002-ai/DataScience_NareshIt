# Scatter Plot:
- A scatter plot is used to show the relationship between two numerical variables.

- scatter plot in seaborn library
- Identify correlation
- Detect Outliers
- Find clusters
- Understand Trends

# When to Use Scatter Plot?

-  Relationship between two numeric variables
-  Before applying Linear Regression
-  Correlation analysis
-  Feature selection in ML
-  Checking pattern & clusters

```python
# scatter plt: total bills vs tip,with hue and size
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time',size='size',palette='deep')
plt.title('Total bill vs tips')
plt.show()
```

# Line Plot:
- A line plot is used to show the trend of data over time or ordered categories.

helps to
- identify trends 
- compare changes over time
- detect seasonality
- analyze growth or decline

# When to Use Line Plot?

-  Time-series data
-  Trend analysis
-  Stock prices
-  Sales over months
-  Temperature changes
-  Performance metrics

```python
# total bill by party size ,split by sex
sns.lineplot(data=tips,x='size',y='total_bill',hue='sex',ci=None,marker='o')
plt.title('total bill by party size and sex')
plt.show()
```

# Bar Plot:
- A bar plot is used to compare values across categories.
- Each category is shown as a bar, and the height/length = value.

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',palette='muted')
```

# Box Plot
- A box plot shows the distribution of data and helps detect outliers.

- Instead of showing every point, it summarizes the data using 5 quatiles. 

```python
# total bill by day split by time divide by smoker
sns.boxplot(data=tips,x='day',y='tip',hue='smoker',palette='Set1') # palette has different color sets
```

# Violin PLot
- A violin plot is an advanced statistical visualization that combines:

- Box Plot (summary statistics) + KDE Density Plot (data shape)

- So instead of only knowing median and quartiles, you understand how the data is distributed internally.

 What Actually Happens Inside

- Seaborn calculates a Kernel Density Estimation (KDE).

- KDE converts raw numbers  smooth probability curve.

- So the violin shape width represents:

- Wider = many observations
- Narrow = few observations
         

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='time',split=True,palette='Set2')
plt.title('vilion plot: total bill by day and time')
```

# Count Plot
- A count plot shows the number of observations in each category.

- It answers: How many?
- Not average, not sum  just frequency (counts)

```python
# orders by day split by smoker
sns.countplot(data=tips,x='day',hue='smoker',palette='dark')
```

# Regression Plot
- A regression plot visualizes the relationship between two numerical variables by drawing:

Scatter points + Best-fit prediction line


```python
sns.regplot(data=tips,x='total_bill',y='tip',scatter_kws={'s':50},line_kws={'color':'red'})
```

# Histogram (Distribution Plot): Total bill with kde
- A histogram shows how a numerical variable is distributed by grouping values into ranges called bins.

It answers: Where most values lie?

```python
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='green')
plt.show()
```

# Pair Plot
- Relationship b/w numerical variable

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
```

# Cat Plot (Point Plot)
- catplot() is a high-level function that can create many categorical graphs in one command.

- Think of it as a wrapper that can draw
- barplot, countplot, boxplot, violinplot, stripplot, swarmplot, pointplot

Unlike sns.barplot()  single graph
sns.catplot()  multiple graphs (grid/facets)

```python
sns.catplot(data=tips,x='day',y='tip',hue='sex',kind='point',palette='bright')
```

# Joint Plot:
- jointplot() is a figure-level Seaborn plot that shows:

- Relationship between two variables + Distribution of each variable in one visualization

- So instead of drawing scatter + histogram separately, it combines them.

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='smoker',palette='coolwarm')
```

# FacetGrid
- FacetGrid lets you split your dataset into subsets and draw the same plot for each subset automatically.

- Instead of filtering data again & again  Seaborn does it for you.

```python
g=sns.FacetGrid(tips,col='time',row='smoker',margin_titles=True,palette='Set2')
g.map(sns.histplot,'total_bill',bins=15)
g.fig.suptitle('FacetGrid: Total Bill by Time and Smoker',y=1.0)
plt.show()
```

# Strip Plot
- A strip plot displays all individual observations of a categorical variable.

- Unlike bar/box/violin  it does NOT summarize
- It shows the actual raw data points


**So you can see:**

 Exact values

Spread

Clusters

Outliers

```python
sns.stripplot(data=tips,x='day',y='tip',hue='sex',palette='Set1',jitter=True)
```

# KDE Plot (Kernel Density Plot)
- A KDE plot shows the true probability distribution of a numerical variable using a smooth curve.

- Think of it as a smoothed histogram

- Instead of bars  continuous density line

```python
sns.kdeplot(data=tips,x='total_bill',hue='sex',fill=True,palette='Set1')
```
