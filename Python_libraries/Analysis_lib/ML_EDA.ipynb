     - non -relavant attribute

3. Bivarate Analyis
     - 2 variable analysi
     - correlation
       1. Positive corr (range(0 to 1)
       2. negetive  (-1 to 0)
       3. no corr (0)
    - range the corr is -1 to 1

     - every time we need to consider +ve corr
4. Outlier Treatment
   - the value very far from the other observation
     
5. Missing Value Treatment
    **how to fill missing values in data**
   - if numerical data then fill with central tendency(mean,median)

   - if categorical data is missing value fill with (mode)
  
   - if mode is have mutliple  value then fill with  nearest neaghbour columns(knn)
  

6. Variable Imputation
   -  imputation also called  (Transformer)
   -  it will convert catogirical value into numerical value
      **Technique**
      - OHE (one hot encoding): it create columns for each unique value
      - dummy model : binary value
      - label encoder: convert data into numical data
