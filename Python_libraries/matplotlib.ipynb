# State Machine Plot

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quardratic')
plt.plot(x,x**3,label='cubic')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title('State machine Plot')

plt.legend()
plt.show()
```

# Working with Numpy array

```python
# evenly sampled time at 200ms interval
t=np.arange(0.,5.,0.2)

# red dashes,blue squares and green triangles
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
```

# Object Oriented API

```python
# fisrt create  a grid of plots
# ax will be an array of two Axes objects
fig,ax=plt.subplots(2)

# call plot() method on the appropriate object
ax[0].plot(x1,np.sin(x1),'b-')
ax[1].plot(x1,np.cos(x1),'r-')
```

# Objects and reference

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');
```

# Multiline Plots

```python
x4 = range(1, 5)

plt.plot(x4, [xi*1.5 for xi in x4])

plt.plot(x4, [xi*3 for xi in x4])

plt.plot(x4, [xi/3.0 for xi in x4])

plt.show()
```
