```python
id(mylist),id(mylist1) # the address of both mylist &mylist1 will be same
```

```python
mylist2=mylist.copy() # create a shallow copy of the list.
```

```python
mylist[0]=1
```

# ALL/ANY
# the all()method returns:
# .True- if all elements a list are true
# .False-if any elements in  a list is false
# the any() function returns True if any element in the list is True.If not,any() return False

```python
l1=[1,2,3,4,0]
```

```python
all(l1) #one value is false
```

```python
any(l1)
```

```python
l2=[1,22,32,43,'True',False]
```

```python
all(l2) 
```

```python
any(l2)
```

```python
l3=[2,32,4,12,"jnjn",True]
all(l3)
```

```python
any(l3)
```
