# Filtering A DataFrame

```python
ipl.head()
```

```python
# find the all final winners
final_match=ipl['MatchNumber']=='Final'
new_df=ipl[final_match]
new_df[['Season','WinningTeam']]
```

```python
# matches with super over in ipl
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# csk won in kolkata
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam']=='Chennai Super Kings')].shape[0]
```

```python
# toss winner and match winner
round((ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0])/(ipl.shape[0]),2)*100
```

```python
# movies with rating>8 and votes>10000
movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)].shape[0]
```

```python
# Action movies with rating>7.5
mask1=movies['genres'].str.split('|').apply(lambda x:'Action' in x)
```

```python
# or
mask1=movies['genres'].str.contains('Action')
```

```python
mask2=movies['imdb_rating']>7.5
movies[mask1&mask2].shape[0]
```
