# BroadCasting
- Used in Vectorization

- The term broadcasting describes how NumPy treats arrays with different shapes during **arithmatic operations**.

- The smaller array is 'Broadcast' across the larger array so that they have compatible shapes.

- note - broadcasting is like matrix operations in math but same dimesions

```python
# same shape
a=np.arange(6).reshape(2,3)
b=np.arange(6,12).reshape(2,3)

print(a)
print(b)
print(a+b)
```

```python
# diff shape
a=np.arange(6).reshape(2,3)
b=np.arange(3).reshape(1,3)
print(a)
print(b)

print(a+b)
```

```python
# more example
a=np.arange(12).reshape(4,3)
b=np.arange(4).reshape(2,2)
a
```

```python
import numpy as np
a = np.array([1, 2, 3])           # shape (3,)
b = 10                            # scalar  shape (1,)
print(a + b)                      # [11 12 13], b broadcasts across a [web:74]

a = np.array([[1,2,3], [4,5,6]])  # (2,3)
b = np.array([10,20,30])          # (3,)  prepends to (1,3)
print(a + b)                      # [[11 22 33] [14 25 36]] [web:77]

a = np.ones((3,3))                # (3,3)
b = np.arange(3)                  # (3,)  (1,3)
print(a + b)                      # Adds row-wise [web:73]
```

- These arrays, xv and yv, each seperately give the x and y coordinates on a 2D grid. You can do

xv**2+yv**2

- This can be done on a larger scale to plot surface plots of 2D functions 
Generate functions f(x, y) = e(x+y) for 2  x  2 and 1  y 1 

```python
x=np.linspace(-2,2,100)
y=np.linspace(-1,1,100)
xv,yv=np.meshgrid(x,y)
f=np.exp(-xv**2-yv**2)
```

Note: pcolormesh is typically the preferable function for 2D plotting, as opposed to imshow or 
pcolor, which take longer.) 

```python
plt.figure(figsize=(6,3))
plt.pcolormesh(xv,yv,f,shading='auto')
plt.colorbar()
plt.grid()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
 
def f(x, y): 
  return np.where((x**2 + y**2 < 1), 1.0, 0.0) 
 
x = np.linspace(-5, 5, 500) 
y = np.linspace(-5, 5, 500) 
xv, yv = np.meshgrid(x, y) 
rectangular_mask = f(xv, yv) 
 
plt.pcolormesh(xv, yv, rectangular_mask, shading='auto') 
plt.colorbar() 
plt.grid() 
plt.show()
```
