# Numpy Array Vs Python Sequences
- Numpy arrays have a fixed size at creation ,unlike python lists(which can grow dynamically).
changing the size of an ndarray will create a new array and delete the original.

- The elements in a Numpy array are all required to be of the same data type and thus will be the same size in memory

- Numpy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically ,such operations are executed more efficiently and with less code than is possible using Python's built-in sequences.

- A growing plethora of scientific and mathematical Python -based packages ara using Numpy arrays; though these typiocally supports python -sequence input, they convert such input to Numpy arrays prior to processing and they often output Numpy arrays.

# round/floor/ceil

# 1.round()
- The numpy.round() function rounds the elements of an array to the nearest integer or to the specified number of decimals.

```python
# round to the nearest integer
arr=np.array([1.2,34.3,12.2,3.7,9.8])
rounded_arr=np.round(arr)
print(rounded_arr)
```

```python
# Round to two decimals
arr=np.array([1.234,2.332,13.423])
np.round(arr,decimals=2)
```

```python
# rondomly
np.round(np.random.random((2,3))*100)
```

# floor
- the numpy.floor() function return the smallest integer less than or equal to each element of an array.

```python
# floor operations
arr=np.array([1.2,2.7,3.5,49.3]) # compact the float to lower intiger
floored_arr=np.floor(arr)
print(floored_arr)
```

```python
np.floor(np.random.random((2,3))*100)
```

# ceil
- The numpy.ceil() function returns the largest integer greater than equal to each element of an array 

```python
arr=np.array([1.2,2.4,3.4,4.9])
ceiled_arr=np.ceil(arr)
print(ceiled_arr)
```
