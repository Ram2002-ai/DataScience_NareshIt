# copy and view
- View → shares memory with original data

- Copy → independent data in memory

- In pandas, Series often becomes a VIEW when created from a DataFrame.

```python
import pandas as pd

df = pd.DataFrame({
    'A': [10, 20, 30, 40],
    'B': [1, 2, 3, 4]
})

s = df['A']     # ⚠️ VIEW
```

```python
s.loc[0] = 999
```

```python
df # original dataframe is changed
```

```python
# copy
s_copy = df['A'].copy()
```

```python
s_copy.loc[0] = 500
```

```python
df # copy does not change the original dataframe
```
