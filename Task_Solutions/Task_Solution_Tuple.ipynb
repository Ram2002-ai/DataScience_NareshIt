# Tuple Slicing

```python
mytuple=('one','two','three','four','five','six','seven','eight')
```

```python
mytuple[0:3]# return all items from 0-2  index location
```

```python
mytuple[2:5] # return all items from 2-4 index location
```

```python
mytuple[:3] #return first three index items 
```

```python
mytuple[-3:]#return last three items
```

```python
mytuple[-2:]
```

```python
mytuple[-1] #return last index item
```

```python
mytuple[:] # return whole tuple
```
