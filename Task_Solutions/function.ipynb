# Function Arguments
- Arguments are the values passed inside the parenthesis of the function.

-   A function can have any number of arguments separated by a comma.



```python
def evenOdd(x):
    if (x%2==0):
        return 'Even'
    else:
        return 'Odd'
print(evenOdd(4))
print(evenOdd(19))
```

# 4. Arbitrary Argument:
- In python Arbitrary Keyword Arguments,*arg and **kwargs  can pass a variable number of arguments to a function using special symbols.There are two special symbols:
- *args in Python (Non-keyword Argument) also called variable length argumnt
- *args store value in tuple

- **kwargs store data in the dictionary 

- **kwargs in Python (keyword Argument) must need to pass the keyword in actual argument

```python
def myfunc(*args):
    print('Variable length argument (*arg):')
    for arg in args:
        print(arg)
myfunc(3,2.3,2,1,2)
```

```python
def myfunc(a,*args):
    print('Variable length argument (*arg):')
    for arg in args:
        print(arg)
    print('a:',a)
myfunc(3,2.3,2,1,2)
```

```python
def myfunc(a,*args):
    print(a+args) # args in tuple 
    print('Variable length argument (*arg):')
    for arg in args:
        print(arg)
    print('a:',a)
myfunc(3,2.3,2,1,2)
```

```python
def myfunc(a,*args):
    print('Variable length argument (*arg):')
    c=a
    for arg in args:
        print(f'{c}+{arg}={c+arg}')
        c=c+arg
       
    print('a:',a)
myfunc(3,2.3,2,1,2)
```

```python
# **kWargs
def myfunc(**kwargs):
    print('keyword argument(**kwargs):')
    for  key,value in kwargs.items():
        print(f'{key}=={value}')
myfunc(first='1',mid='50',last='100')
```

```python
# **kWargs
def myfunc(**kwargs):
    print('keyword argument(**kwargs):')
    for  key,value in kwargs.items():
        print(f'{key}=={value}')
myfunc('hey','welcome') #**kwargs only accept values with keywords
```

```python
def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

# Function call with both types of arguments
myFun('Hey', 'Welcome', first='Geeks', mid='for', last='Geeks')
```

# Function within Functions
- A function defined inside another function is called an **inner function** (or nested function).

- It can access variables from the enclosing function's scopes and is often used to keep logic protected and organized.

```python
def f1():
    s='I love <my love>'
    def f2():
        print(s) #accessing the s variable from f1()
    f2() # second function calling
f1()
```

# Recursive Functions:
- a recursive function is a function that calls itself to solve a problem.

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(6))
```
