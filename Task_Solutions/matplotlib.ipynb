```python
batsman=pd.read_csv(r"E:\data\matplotlib_data\sharma-kohli.csv")
batsman.head()
```

```python
# categorical - numerical
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# ploting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'],label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],label='rohit')

plt.title('Rohit Vs Virat Career Comparision')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend()
plt.show()
```

# colors (hex) and line (width and style) and marker(size)

```python
# colors (hex) and line (width and style) and marker(size)

plt.plot(batsman['index'],batsman['V Kohli'],label='virat',color='#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],label='rohit',color='#FC00D6')

plt.title('Rohit Vs Virat Career Comparision')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend()
plt.show()
```

```python
# colors (hex) and line (width and style) and marker(size)
# here ls for (linestyle) and lw for ()
plt.plot(batsman['index'],batsman['V Kohli'],label='virat',color='#D9F10F',ls='solid',lw=3)
plt.plot(batsman['index'],batsman['RG Sharma'],label='rohit',color='#FC00D6',ls='dashdot',lw=2)

plt.title('Rohit Vs Virat Career Comparision')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend()
plt.show()
```

# marker and marker size

```python
# marker and marker size
# colors (hex) and line (width and style) and marker(size)

plt.plot(batsman['index'],batsman['V Kohli'],label='virat',color='#D9F10F',marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],label='rohit',color='#FC00D6',marker='o')

plt.title('Rohit Vs Virat Career Comparision')
plt.xlabel('Season')
plt.ylabel('Run Scored')
plt.legend()
plt.show()
```

# Bar Chart
- Bivariate Analysis
- Numerical vs Categorical

```python
# simple bar chart
children=[10,20,40,10,30]
colors=['red','blue','green','yellow','pink']
plt.bar(colors,children,color=colors)
```

```python
data = pd.DataFrame({
    "Category": ["A", "B", "C", "D"],
    "Value": [10, 20, 15, 25]
})

sns.barplot(x="Category", y="Value", data=data, hue="Category", palette="Set2", legend=False)

plt.show()
```

```python
df=pd.read_csv(r"E:\data\matplotlib_data\batsman_season_record.csv")
```

```python
df
```

# multiple bar chart/xticks

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

# Stacked Bar chart

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
