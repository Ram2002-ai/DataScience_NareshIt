# Copy Set

```python
myset={'one','two','three','four','five','six','seven','eight'}
myset
```

```python
myset1=myset #create a new reference it allocated same memory
myset1
```

```python
id(myset),id(myset1),len(myset) # the address of both myset & myset1 will be same as
# we also run multiple function in same line
```

```python
my_set=myset.copy() #create a copy of the set
my_set # it allocated different memory location
```

```python
id(my_set)
```

```python
print(myset)
print(my_set)
```

# Difference

```python
A={1,2,3,4,5}
B={4,5,6,7,8}
```

```python
A.difference(B) # sets of element which is only in A not in B (B's all' elements remove)
```

```python
B-A #set of element which is only in B not in A (A's all elements remove)
```

# difference_update()

```python
"""
Updates the set calling the difference_update() method with the difference of set
For below example Set B will be updated with the difference of B&A
"""
B.difference_update(A)
```

```python
print(A)
print(B)
```

# Subset,Superset & Disjiont

```python
A={1,2,3,4,5,6,7,8,9}
B={3,4,5,6,7,8}
C={10,20,30,40}
```

```python
B.issubset(A) # B is subset of A
```

```python
A.issuperset(B) # A is superset of B
```

```python
C.isdisjoint(A) # C is neighbour of A (because no common elements in both)
```

```python
B.isdisjoint(A) # not disjoint because common element present in both
```

# All/Any
**the all() method returns**
**True - if all keys of the dictionary are true**
**False-if any keys of the dictionary is false**
**the any() returns True if any key of the dictionary is True.if not any() return false**



```python
mydict
```

```python
all(mydict) # all true value present
```

```python
mydict[0]='h'
```

```python
all(mydict) # atleast 1 false value is present in the dictionary
```

```python
any(mydict) # atleast 1 true items is present in the mydict
```
