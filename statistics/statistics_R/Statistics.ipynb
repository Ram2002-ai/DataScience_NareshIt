# Descriptive Statistics

# 1. Measure of Central  Tendency

```python
# data creation
import pandas as pd
import numpy as np

# sample order value data (in INR)
order_values=[
    499,799,999,1499,1999,2499,2999,3499,3999,599,899,1299,1799,2999,3499,4999,6999
    ,9999,1599,1899,2199
]

df=pd.DataFrame({'Order_values':order_values})
df.head()
```

# 1. Mean (Arithmatic Average)

```python
mean_value=df['Order_values'].mean()
round(mean_value)
```

# 2. Median (Arithmatic median)
- Means sorted center value of numerical data
- if data is in odd number then median is meddle value
- if data is in even number then n/2+1 is the middle value

```python
median_value=df['Order_values'].median()
median_value
```

# 3. Mode (most frequent value)

```python
mode_value=df['Order_values'].mode()
mode_value
```

# 2. Measure of Dispersion

# 4. Variance

```python
variance=df['Order_values'].var()
round(variance,2) # because units are squared thats why it not give a proper result
```

# 5.Standard Deviation(SD)
- Average deviation from the mean

```python
std_dev=df['Order_values'].std()
round(std_dev,2)
```

# Interquartile Range(IQR)

```python
Q1=df['Order_values'].quantile(.25)
Q3=df['Order_values'].quantile(.75)
Q2=df['Order_values'].quantile(.50)
minimum=Q1-1.5*IQR
maximum=Q3+1.5*IQR
IQR=Q3-Q1
Q1,Q3,IQR,Q2,minimum,maximum
```

# 7. Mean Absolute Deviation (MAD)

```python
mad=df['Order_values'].mad()
mad
```

# 3. Shape of Data Distribution


# 8. Skewness


```python
# 8. Skewness
skewness=df['Order_values'].skew()
round(skewness,2)
```

# 9. Kurtosis (tells peakness of data)

```python
kurt=df['Order_values'].kurtosis()
round(kurt,2)
```

# 4. Percentiles , Deciles (Decision Metrix)

# 10. Percentiles

```python
df['Order_values'].quantile([0.1,0.25,0.5,0.75,0.9])
```

# 11. Deciles

```python
deciles=df['Order_values'].quantile(np.arange(0.1,1.1,0.1))
deciles
```

# Outlier Detection Techniques
 1. IQR
 2. Z- Score

# IQR

```python
outlier=df[df['Order_values']>maximum ]
outlier
```

```python
min_outlier=df[df['Order_values']<minimum]
min_outlier # no values less than minimum
```

# Z-Score

```python
from scipy.stats import zscore
```

```python
df['z_score']=zscore(df["Order_values"])
df.head()
```

```python
df[(df['z_score']>3) | (df['z_score']<-3)]
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n=5000

data = pd.DataFrame({
    'Order_Id':np.arange(1,n+1),
    'Customer_Age':np.random.randint(18,65,n),
    'Gender':np.random.choice(['male','female'],n),
    'City':np.random.choice(['Hydrabad','Banglore','Chennai','Mumbai','Delhi'],n),
    'Product_Category':np.random.choice(
        ['Electronics','Fashion','Home','Sports','Beauty'],n),
    'Price':np.random.randint(200,20000,n),
    'Discount (%)':np.random.randint(5,40,n),
    'Rating':np.round(np.random.uniform(1,5,n),1)
})
data
```
